```python
import pandas as pd
from pyecharts import Pie  
```

# 数据处理

```python
data = pd.read_csv('./data_processed.csv')
data
```

```python
data = data.fillna(method='ffill')
```

```python
age = data['agejoin']
age
```

```python
age.describe()
```

```python
c_20 = 0
c_20_30 = 0
c_30_40 = 0
c_40_50 = 0
c_50_60 = 0
for i in age:
    if i<=20:
        c_20+=1
    elif i<=30:
        c_20_30+=1
    elif i<=40:
        c_30_40+=1
    elif i<=50:
        c_40_50+=1
    elif i<=60:
        c_50_60+=1
```

```python
attr = ["age<=20","20<age<=30","30<age<=40","40<age<=50",'50<age<=60']
v1 = [c_20,c_20_30,c_30_40,c_40_50,c_50_60]
pie = Pie("Terroist'age",title_pos='center',width=900)
pie.add("Terroist'age",attr,v1,center=[50,50],is_random=True,radius=[30,75],rosetype='area',is_legend_show=False,is_label_show='True')
pie
```
